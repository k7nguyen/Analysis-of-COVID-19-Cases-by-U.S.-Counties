```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loads data and shows first n rows to test if object has correct data
df=pd.read_csv("us-counties.csv", index_col="date")
df.head()
```

```python
#Column names
df.columns
```

# California: Los Angeles County

```python
#Shows earliest and latest data for LA County cases
la_df=df[df.county=="Los Angeles"]
la_df
```

```python
#Line chart of LA County cases from Janaury until now (early June)
ax=la_df.plot(y="cases", figsize=(10,6))
ax.set(xlabel="Date", ylabel="Cases")
plt.title("COVID-19 Cases in Los Angeles County from January to Early June")
```

#### This graph displays the overall trend of COVID-19 cases in Los Angeles County. Although the chart shows that the first case in this county occured in late January, the graphs shows that the the number of cases started to increase in late March. Since then, the number of cases have been rapidly increasing and as of early June, the number of cases in LA County is roughly 65,000. An interesting observation seen in this graph is that the county's Shelter-in-Place orders do not seem to have an extremely large negative effect on the number of COVID-19 cases. It is likely that this order kept the number of cases lower than if the order was not implemented, but it has not entirely stopped the virus from spreading.

# Washington: Spokane County & Snohomish County

```python
#Shows earliest and latest data for Washington state cases
washingtondf=df[df.state=="Washington"]
washingtondf
```

```python
#Shows earliest and latest data for Spokane County cases
spokanedf=washingtondf[washingtondf.county=="Spokane"]
spokanedf
```

```python
#Shows earliest and latest data for Snohomish County cases
snohomishdf=washingtondf[washingtondf.county=="Snohomish"]
snohomishdf
```

```python
#Line chart of Spokane and Snohomish cases from Janaury until now (early June)
dataframe={"date": snohomishdf.index, "Spokane": [], "Snohomish": []}

for i in snohomishdf.index:
    if i in spokanedf.index:
        dataframe["Spokane"].append(spokanedf.loc[i, "cases"])
    else:
        dataframe["Spokane"].append(0)
        
    dataframe['Snohomish'].append(snohomishdf.loc[i]['cases'])

graphdf=pd.DataFrame.from_dict(dataframe)

ax=graphdf.plot(x="date", y=["Spokane", "Snohomish"], figsize=(10,6))
ax.set(xlabel="Date", ylabel="Cases")
plt.title("COVID-19 Cases in Spokane & Snohomish from January to Early June")
```

#### This graph shows the number of cases in Spokane compared to the number of cases in Snohomish from January to early June. In  early/mid March, Snohomish began to report a surplus of cases, while Spokane remained constant until early April. Since then, it seems that Snohomish continues to report throusands more cases than Spokane. However, both seem to have an upward trend.
